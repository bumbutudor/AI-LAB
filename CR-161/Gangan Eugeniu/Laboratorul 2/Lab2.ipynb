```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
%matplotlib inline
```

```python
neww_csv = pd.read_csv("imbd_superhero.csv", 
                  header=None, 
                  names=['An', 'Titlu','Organizatia', 'IMBD', 'AltRating', 'DataX', 'UnknownColumn', 'PretMediulaBilet', 'NrDeSpectatori', 'TopPentruAncutare'])

altrating_norm = np.where(neww_csv['IMBD']>=7, 'yes', 'no')
neww_csv.insert(10, 'StatusPrivit', altrating_norm)
neww_csv.head(8)
```

```python
dc = neww_csv[neww_csv['Organizatia']=='DC']
dc.head(7)
```

```python
marvel = neww_csv[neww_csv['Organizatia']=='Marvel']
sarc4=pd.DataFrame(marvel)
sarc4=sarc4[['An','Titlu']]
sarc4
```

```python
new_csv = pd.read_csv("imbd_superhero.csv", 
                  header=None, 
                  names=['An', 'Titlu','Organizatia', 'IMBD', 'AltRating', 'DataX', 'UnknownColumn', 'PretMediulaBilet', 'NrDeSpectatori', 'TopPentruAncutare'])

plot = new_csv.plot.scatter(x = 'An', y = 'IMBD')
z = np.polyfit(x=new_csv.An, y=new_csv.IMBD, deg=5) #polinom de gradul 5 inseamna deg=5
p = np.poly1d(z)
trendline = pd.DataFrame(data=p(new_csv.An), index=new_csv.An)
trendline.plot.line(ax=plot, color='Black')
```
