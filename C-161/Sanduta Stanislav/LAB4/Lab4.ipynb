```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
student_data = pd.read_csv('D:\Anul IV\AI\Laboratorul II\student-mat.csv', header = 1, delimiter=';')
student_data.columns = ['school', 'sex', 'age', 'address', 'famsize', 'pstatus', 'medu', 'fedu', 'mjob', 'fjob',
                        'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup',
                        'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'farmel', 'freetime', 'goout',
                        'dalc', 'walc', 'health', 'absences', 'g1', 'g2', 'g3']
student_data
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder

enc = preprocessing.OrdinalEncoder()

enc.fit(student_data)

data = enc.transform(student_data)

student_data = pd.DataFrame({'school': data[:, 0], 'sex': data[:, 1], 'age': data[:, 2], 'address': data[:, 3],
                             'famsize': data[:, 4], 'pstatus': data[:, 5], 'medu': data[:, 6], 'fedu': data[:, 7],
                             'mjob': data[:, 8], 'fjob': data[:, 9], 'reason': data[:, 10], 'guardian': data[:, 11],
                             'traveltime': data[:, 12], 'studytime': data[:, 13], 'failures': data[:, 14],
                             'schoolsup': data[:, 15], 'famsup': data[:, 16], 'paid': data[:, 17], 'activities': data[:, 18],
                             'nursery': data[:, 19], 'higher': data[:, 20], 'internet': data[:, 21], 'romantic': data[:, 22],
                             'farmel': data[:, 23], 'freetime': data[:, 24], 'goout': data[:, 25], 'dalc': data[:, 26],
                             'walc': data[:, 27], 'health': data[:, 28], 'absences': data[:, 29], 'g1': data[:, 30],
                             'g2': data[:, 31], 'g3': data[:, 32]})

student_data['studytime']
```

```python
features = ['studytime', 'romantic', 'absences', 'health']
X = student_data[features]
X = X[X['absences'] < 20]
X = X[X['studytime'] > 0]
Y = student_data[student_data['absences'] < 20]
Y = Y[Y['studytime'] > 0]
Y = Y.g3
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.03, random_state=1)
X_test
```

```python
classifier = DecisionTreeClassifier(criterion="entropy")
classifier_decision_tree = classifier.fit(X_train,Y_train)
```

```python
Y_predicted = classifier_decision_tree.predict(X_test)

print("Accuracy:",metrics.accuracy_score(Y_test, Y_predicted))
print(Y_test, Y_predicted)
```

```python
prediction = X_test[X_test["studytime"] == 3.0]
classifier_decision_tree.predict(prediction)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

classes = ["%.2f" % number for number in classifier_decision_tree.classes_]

dot_data = StringIO()
export_graphviz(classifier_decision_tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = features,class_names=classes)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('final_mark.png')
Image(graph.create_png())
```
